```python
import requests
from bs4 import BeautifulSoup
r = requests.get("https://www.scan.co.uk/shop/computer-hardware/laptops-gaming/3063/2835/3064/2836/2837", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c=r.content
soup= BeautifulSoup(c,"html.parser")
all=soup.find_all("li",{"class":"product"})
all[0].find("span",{"class":"price"}).text
```

```python
l=[]
for item in all:
    d={}
    try:
        d["Price"]=item.find("span",{"class":"price"}).text
    except:
        d["Price"]=None
        
    try:
        d["Description"]=item.find("span",{"class":"description"}).text
    except:
        d["Description"]=None
        
    try:
        d["availability"]=item.find("a",{"class":"btn"}).text
    except:
        d["availability"]=None
        
    try:
        d["Link No"]=item.find("span",{"class":"linkNo"}).text
    except:
        d["Link No"]=None
        
    print(" ")
    l.append(d)
```

```python
import pandas
df=pandas.DataFrame(l)
```
